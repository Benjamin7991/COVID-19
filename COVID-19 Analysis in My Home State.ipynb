```python
import pandas as pd
import os
import datetime 
```

# Lets first compile all data into single df

```python
#df = pd.read_csv('./csse_covid_19_daily_reports_us/09-04-2020.csv')

#files = [file for file in os.listdir('./csse_covid_19_daily_reports_us')]

files=[f for f in os.listdir("./csse_covid_19_daily_reports_us") if f.endswith('.csv')]
all_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('./csse_covid_19_daily_reports_us/' + file)
    all_data = pd.concat([all_data, df])
    


all_data.tail()
```

## Combined all files in a dir into one csv file. Then, filtered out my state.

```python
all_data = all_data[all_data['Province_State'] == 'Colorado']
```

```python
all_data.head()
```

## Combined all files in a dir into one csv file. Then, filtered out my state.

```python
del all_data['Long_']
del all_data['Lat']
del all_data['UID']
del all_data['FIPS']
del all_data['Country_Region']
del all_data['ISO3']
```

#### Deleted longtitude and latitude..^

```python
all_data['Last_Update'] = pd.to_datetime(all_data['Last_Update'])
```

```python
all_data['Month'] = all_data['Last_Update'].dt.month
```

### Grouped by month, will display cases over time with this ^

```python
all_data.head()
```

```python
#all_data.groupby(all_data['Month'])
all_data = all_data.sort_values(by = 'Deaths', ascending=True)
```

### lets graph changes in death over time

```python
all_data.head(50)
```

```python
import matplotlib.pyplot as plt


plt.plot(all_data['Last_Update'], all_data['Deaths'])
plt.xlabel('Date', size = 15)
plt.ylabel('Deaths', size = 15)
plt.title('Covid-19 Deaths in Colorado Over Time', size = 15)
plt.xticks(size = 10, rotation= 65)
plt.yticks(size = 10)


plt.show()
```
